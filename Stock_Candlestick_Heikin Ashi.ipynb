```python
import pandas as pd
import yfinance
from mpl_finance import candlestick_ohlc
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = [15,10]
plt.rc('font',size=18)
```

```python
## Enter the stock/index name in ticker_name. 
##Choose start and end date in YYYY-MM-DD format. 
##Choose the interval.
```

```python
ticker_name = 'SBIN.NS'
t = yfinance.Ticker(ticker_name)
df = t.history(interval='1wk',start='2021-01-01',end='2021-09-25')
df['Change %'] = ((df['Close']-df["Open"])/df['Close'])*100
df.reset_index(inplace=True)
df = df.loc[:,['Date','Open','High','Low','Close',"Change %"]]
```

```python
df.tail()
```

```python
## Using plotly to create a function for creating charts
```

```python
import plotly.graph_objects as go
def chart(df):
    fig = go.Figure(data=[go.Candlestick(x=df['Date'],
                                        open=df['Open'],
                                        high = df['High'],
                                        low = df['Low'],
                                        close = df['Close'])]) 
    fig.update_layout(
        title=ticker_name,
        yaxis_title='Price',
        width=1000,
        height=1000)
    fig.show()
```

```python
chart(df)
```
